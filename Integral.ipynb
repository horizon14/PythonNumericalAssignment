### Create a python function that computes $ f(x) = e ^{âˆ’4 sin(2\pi x)}$ and then plot the function from $ x = 0 $ to $x = 1$

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.exp(-4)*np.sin(2*np.pi*x)

x=np.linspace(0,1,1000)
y=f(x)

plt.plot(x,y)
plt.title("Plot of f(x) = eâˆ’4 sin(2Ï€x)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()
```

### One way of computing the integral of a function numerically is using the approximation $ \sum_i f(x_i) \Delta x $ where the sum is only for $  x_i $ between the bounds of integration.
#### Use the np.sum function and the spacing between the values in the x array (i.e. âˆ†x) from question 1 to evaluate $ \int_0^1 e^{ -4 sin(2\pi x)}dx $

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.exp(-4)*np.sin(2*np.pi*x)

x=np.linspace(0,1,1000)

y=f(x)

delta_x=x[1]-x[0]

plt.plot(x,y)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

integral_xy=np.sum(y*delta_x)

print("The integral of f(x) = eâˆ’4 sin(2Ï€x) from 0 to 1 is", integral_xy)
```

#### The numpy function np.cumsum(arr) computes the cumulative sum of elements in an array arr. For example, if arr = [ 0,4,7,2]ï¼Œthen np.cumsum(arr)= [ 0,4,11,13]. Use this function to compute and plot $ F(x)=\int_0^x e^{-4sin(2\pi x')}dx' $

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.exp(-4)*np.sin(2*np.pi*x)
x=np.linspace(0,1,1000)
y=f(x)

delta_x=x[1]-x[0]
F=np.cumsum(y*delta_x)

plt.plot(x,F)
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("F(x)")
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend(["F(x)","f(x)"])
plt.show()
```

### One way to compute the derivative of a function numerically is using the approximation $ fâ€² = âˆ†fi/âˆ†x$
#### The np.diff function is used to compute the differences between consecutive elements of an array. For example, if arr = [ 0,4,7,2] then np.diff(arr)= [ - 4,3,-5 ] (Note: The length becomes 1smaller ). Use this to compute âˆ†f from the data in Question 1
#### Compute $fâ€² = âˆ†f/âˆ†x $ and plot as a function of x (Hint: Youâ€™ll need to make sure the x and y arrays youâ€™re plotting are the same size.Check their lengths before plotting)

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.exp(-4)*np.sin(2*np.pi*x)
x=np.linspace(0,1,1000)
y=f(x)

delta_x=x[1]-x[0]
f_diff=np.diff(y)/delta_x

plt.plot(x[0:-1],f_diff)
```

#### Using boolean indexing, plot only these again, but only for values where the cs137 data exceeds 1500. Add a grid, x and y labels,and a title to the plot.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df=pd.read_csv("data/scintillator.csv")
plt.figure(figsize=(10,6))
plt.plot(df['Channel'],df['Cs137'],label='All data')

df_filtered=df[df['Cs137']>1500]
plt.plot(df_filtered['Channel'],df_filtered['Cs137'],label='Filtered data')


plt.title("Scintillator data")
plt.xlabel("Channel")
plt.ylabel("Counts")
plt.legend()
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

plt.show()
```
